### How long does it take to close a deal?

From the first contact date, based on the lead behaviour profile, I would like to know how long the lead decides to become Olist's seller.

lead behivour profile

- Shark : Dominance
- Eagle : Influence
- Cat : Steadiness
- Wolf : Conscientiousness

```python
query = """
        SELECT lead_behaviour_profile
            ,round(avg(date_part('day', age(won_date, first_contact_date)))::decimal,2)
             as avg_days_diff
        FROM leads l
        JOIN deals d on l.mql_id = d.mql_id
        GROUP BY 1
        ORDER BY 2
        """

select(query)
```

It seems mixed profile leads are much harder to negotiate.

### Merchants conversion rate

The conversion rate reflect the percentage of signed-up merchants that later become Olist sellers. First, let's see the number of leads and closed deals over time.

```python
query = """
        SELECT a.year
            ,num_deals
            ,num_leads
            ,round(num_deals * 1.0 / num_leads * 100, 2) as conversion_rate
        FROM 
        (
            SELECT date_part('year', first_contact_date)::int as year
                 ,count(distinct mql_id) as num_leads
            FROM leads
            GROUP BY 1
        ) a
        JOIN 
        (
            SELECT date_part('year', won_date)::int as year
                 ,count(distinct mql_id) as num_deals
            FROM deals
            GROUP BY 1
        ) b on a.year = b.year
        """

select(query)
```

The conversion rate in 2017 is very low. In 2018, the conversion rate is still low. I would like to see how the conversion rate is different when it's come to different leads traffic source origin.

```python
query = """
        SELECT origin
            ,sum(year_2017) as conversion_rate_2017
            ,sum(year_2018) as conversion_rate_2018            
        FROM
        (
            SELECT a.origin
                ,case when a.year = 2017 
                    then round(num_deals * 1.0 / num_leads * 100, 2) else 0 end as year_2017
                ,case when a.year = 2018
                    then round(num_deals * 1.0 / num_leads * 100, 2) else 0 end as year_2018
            FROM
            (
                SELECT origin
                    ,date_part('year', first_contact_date)::int as year
                    ,count(distinct mql_id) as num_leads
                FROM leads
                GROUP BY 1,2
            ) a
            JOIN
            (
                SELECT origin
                    ,date_part('year', won_date)::int as year
                    ,count(distinct d.mql_id) as num_deals
                FROM deals d
                JOIN leads l on d.mql_id = l.mql_id
                GROUP BY 1,2
            ) b on a.year = b.year and a.origin = b.origin
        ) c
        GROUP BY 1
        ORDER BY 3 desc
        
        """

select(query)
```

Besides the missing value, 'unknown' source has the highest rate. Olist should find out the missing information and focus on these kind of traffic source to see if it can help improve the conversion rate.

```python
query = """
        SELECT month
            ,sr_id
            ,num_deals
        FROM 
        (
            SELECT date_part('month', won_date)::int as month
                ,sr_id
                ,count(distinct mql_id) as num_deals
                ,rank() over(partition by date_part('month', won_date)::int 
                    order by count(distinct mql_id) desc) as sales_rank
               FROM deals
               WHERE date_part('year', won_date) = 2018
               GROUP BY 1,2
        ) a
        WHERE sales_rank = 1
        """

select(query)
```
