### Number of customers and customers growth rate

I would like to know the number of new customers increasing each year.

```python
query = """
        SELECT date_part('year', order_purchase_timestamp)::int AS year
            ,count(distinct customer_unique_id) AS num_customers
        FROM customers c 
        JOIN orders o on c.customer_id = o.customer_id
        GROUP BY 1
        """

select(query)
```

```python
query = """
        WITH num_customers_year as
        (
            SELECT date_part('year', order_purchase_timestamp)::int AS year
                ,count(distinct customer_unique_id) AS num_customers
            FROM customers c
            JOIN orders o on c.customer_id = o.customer_id
            GROUP BY 1
        )
        SELECT year
            ,first_value(num_customers) over (order by year) AS cohort_size
            ,num_customers
            ,round(num_customers / first_value(num_customers) over (order by year) * 100, 2) AS customers_growth
        FROM num_customers_year
        """

select(query)
```

### Customers repeating rate

```python
query = """
        SELECT a.year, c.repeat_customers, a.num_customers,
               round(c.repeat_customers * 1.0 / a.num_customers * 100, 2) AS repeat_rate
        FROM (SELECT date_part('year', order_purchase_timestamp)::int AS year,
                     COUNT(DISTINCT customer_unique_id) AS num_customers             
              FROM customers c JOIN orders o
              ON c.customer_id = o.customer_id
              GROUP BY year) a
        JOIN (SELECT year, COUNT(DISTINCT customer_unique_id) AS repeat_customers
        FROM (SELECT CASE ROW_NUMBER() 
                OVER(PARTITION BY customer_unique_id 
                     ORDER BY order_purchase_timestamp)
           WHEN 1 THEN 'Total'
           ELSE 'Return'
           END AS customer_type, c.customer_unique_id,
           date_part('year', order_purchase_timestamp)::int AS year
        FROM customers c JOIN orders o
        ON c.customer_id = o.customer_id
        ORDER BY customer_unique_id, order_purchase_timestamp) b
        WHERE customer_type = 'Return'
        GROUP BY year, customer_type) c ON a.year = c.year;
        """

select(query)
```
