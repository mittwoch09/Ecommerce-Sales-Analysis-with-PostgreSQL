### On-time delivery rate

Olist requires merchants to maintain their on-time delivery rate at least 97% for the last 30 days to guarantee their performance.

I would like to know the number of sellers who have 97% or higher on-time delivery rate lately. First, let's check if there are any sellers who delivered orders on time each year in 2018.

```python
query = """
        SELECT date_part('month', order_purchase_timestamp)::int as month,
               count(distinct seller_id) as num_sellers
        FROM 
        (
            SELECT oi.seller_id, o.order_purchase_timestamp
            FROM orders o 
            JOIN order_items oi on o.order_id = oi.order_id
            WHERE order_status = 'delivered'
            and date_part('year', order_purchase_timestamp) = 2018
            and date_part('day', age(o.order_estimated_delivery_date, o.order_delivered_customer_date)) >= 0
        ) a
        GROUP BY month
        """

select(query)
```

The nearest month we have is August. The query below calculate the number of sellers who have 97% or higher on-time delivery rate in August 2018.

```python
query = """
        SELECT c.num_sellers, count(b.seller_id) as num_qualified_sellers
               ,count(b.seller_id) * 1.0 / c.num_sellers * 100 as percentage
        FROM 
        (
            SELECT a.seller_id
                ,num_on_time_orders * 1.0 / count(distinct order_id) * 100 as percent_on_time
            FROM order_items
            JOIN 
            (
                SELECT oi.seller_id
                    ,count(distinct o.order_id) as num_on_time_orders
                FROM orders o 
                JOIN order_items oi on o.order_id = oi.order_id
                WHERE order_status = 'delivered'
                and date_part('month', order_purchase_timestamp) = 8
                and date_part('year', order_purchase_timestamp) = 2018
                and date_part('day', age(o.order_estimated_delivery_date, o.order_delivered_customer_date)) >= 0
                GROUP BY 1
            ) a
            on order_items.seller_id = a.seller_id
            GROUP BY a.seller_id, num_on_time_orders
            HAVING num_on_time_orders * 1.0 / count(distinct order_id) * 100 >= 97
        ) b
        CROSS JOIN 
        (
            SELECT count(distinct seller_id) as num_sellers
            FROM orders o 
            JOIN order_items oi on o.order_id = oi.order_id
            WHERE order_status = 'delivered'
            and date_part('month', order_purchase_timestamp) = 8
            and date_part('year', order_purchase_timestamp) = 2018
        ) c
        GROUP BY 1
        """

select(query)
```

In August 2018, among 1261 sellers who have delivered orders to customers, only 135 (10.71%) have 97% on-time delivery rate.

```python
query = """
        SELECT a.seller_id
            ,count(distinct order_id) as num_delivered_orders
            ,num_on_time_orders
            ,num_on_time_orders * 1.0 / count(distinct order_id) * 100 as percent_on_time
        FROM order_items
        JOIN
        (
            SELECT oi.seller_id
                ,count(distinct o.order_id) as num_on_time_orders
            FROM orders o 
            JOIN order_items oi on o.order_id = oi.order_id
            WHERE order_status = 'delivered'
            and date_part('month', order_purchase_timestamp) = 8
            and date_part('year', order_purchase_timestamp) = 2018
            and date_part('day', age(o.order_estimated_delivery_date, o.order_delivered_customer_date)) >= 0
            GROUP BY 1
        ) a
        on order_items.seller_id = a.seller_id
        GROUP BY 1,3
        HAVING num_on_time_orders * 1.0 / count(distinct order_id) * 100 >= 97
        ORDER BY 2 desc
        LIMIT 10
        """

select(query)
```

Here is the top 10 sellers who have the highest number of delivered orders with 97% on-time delivery rate in August 2018.

### How fast is it for a customer to receive an order?

This query calculates the percentage of orders arriving within 2 days, 1 week, 2 weeks, or more than 2 weeks after they are placed.

```python
query = """
        SELECT
            sum(case
                when date_part('day', 
                age(order_delivered_customer_date, 
                    order_purchase_timestamp)) <= 2 then 1 
                end) * 1.0 / count(order_id) * 100 as under_two_days,
            sum(case when date_part('day',
                age(order_delivered_customer_date, 
                    order_purchase_timestamp)) BETWEEN 3 and 5 then 1
                end) * 1.0 / count(order_id) * 100 as in_one_week,
            sum(case when date_part('day', 
                age(order_delivered_customer_date, 
                    order_purchase_timestamp)) BETWEEN 6 and 14 then 1 
                end) * 1.0 / count(order_id) * 100 as in_two_weeks,
            sum(case when date_part('day', 
                age(order_delivered_customer_date,
                order_purchase_timestamp)) > 14 then 1 
            end) * 1.0 / count(order_id) * 100 as more_than_two_weeks            
        FROM orders
        WHERE order_status = 'delivered'
        and date_part('day', age(order_delivered_customer_date, order_purchase_timestamp)) >= 0
        """

select(query)
```

Most orders were delivered in 2 weeks (54.29%).

### Shipping limit date

When an order is approved, Olist platform creates a deadline requiring the merchant to handle the order to Olist logistics partner before that date. 

In the past 30 days, how many merchants meet this requirement?

```python
query = """
        SELECT c.num_sellers, count(b.seller_id) as num_meet_deadline
               ,count(b.seller_id) * 1.0 / c.num_sellers * 100 as percentage
        FROM 
        (
            SELECT a.seller_id
                ,meet_deadline * 1.0 / count(distinct order_id) * 100 as percentage
            FROM order_items
            JOIN 
            (
                SELECT oi.seller_id
                    ,count(distinct o.order_id) as meet_deadline
                FROM orders o 
                JOIN order_items oi on o.order_id = oi.order_id
                WHERE order_status = 'delivered'
                and date_part('month', order_purchase_timestamp) = 8
                and date_part('year', order_purchase_timestamp) = 2018
                and date_part('day', age(shipping_limit_date, order_delivered_carrier_date)) >= 0
                GROUP BY 1
            ) a
            on order_items.seller_id = a.seller_id
            GROUP BY a.seller_id, meet_deadline
        ) b
        CROSS JOIN 
        (
            SELECT count(distinct seller_id) as num_sellers
            FROM orders o 
            JOIN order_items oi on o.order_id = oi.order_id
            WHERE order_status = 'delivered'
            and date_part('month', order_purchase_timestamp) = 8
            and date_part('year', order_purchase_timestamp) = 2018
        ) c
        GROUP BY 1
        """

select(query)
```

We have 1261 merchants who sucessfully delivered orders to their customers in August 2018. 

97.70% of them meet the shipping deadline, but only 10.71% has 97% or more on-time delivery rate.
