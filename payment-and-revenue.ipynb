### Total revenue

To find the total revenue each year, I would examine delivered orders only since they are completed orders and normally full paid. revenue is sum of price column and freight_value column.

```python
query = """
        SELECT date_part('year', order_purchase_timestamp) as year
            ,sum(revenue) as total_revenue
        FROM (SELECT order_id
                  ,sum(price) + sum(freight_value) as revenue
              FROM order_items
              GROUP BY order_id) a
        JOIN orders o on a.order_id = o.order_id
        WHERE order_status = 'delivered'
        GROUP BY 1
        ORDER BY 1
        """

select(query)
```

Assuming that Olist started in 2016, I also want to know the percentage change of total revenue value each year.

```python
query = """
        WITH total_revenue_year as
        (
            SELECT date_part('year', order_purchase_timestamp)::int as year
                ,sum(revenue) as total_revenue
            FROM (SELECT order_id
                      ,sum(price) + sum(freight_value) as revenue
                  FROM order_items
                  GROUP BY order_id) a
            JOIN orders o on a.order_id = o.order_id
            WHERE order_status = 'delivered'
            GROUP BY 1
        )
        SELECT year
            ,total_revenue
            ,coalesce(lag(total_revenue, 1) over (order by year), 0) as previous_total_revenue
            ,coalesce(round((total_revenue * 1.0 / lag(total_revenue, 1) over (order by year))::decimal, 2), 0) as revenue_growth
        FROM total_revenue_year
        """

select(query)
```

We have the top 3 sellers who have the highest total revenue each year as follow:

```python
query = """
        SELECT year, seller_id, total_revenue, value_rank
        FROM
        (
            SELECT year, seller_id
                ,sum(t1.revenue) as total_revenue
                ,rank() over (partition by year order by sum(t1.revenue) desc) as value_rank
            FROM (
                    SELECT order_id, date_part('year', order_purchase_timestamp)::int AS year
                    FROM orders
                    WHERE order_status = 'delivered'
                 ) o
            JOIN (
                    SELECT seller_id, order_id
                        ,sum(price) + sum(freight_value) as revenue
                    FROM order_items
                    GROUP BY seller_id, order_id
                  ) t1 
            on o.order_id = t1.order_id
            GROUP BY year, seller_id
        ) t2
        WHERE value_rank IN (1,2,3)
        """

select(query)
```

### Revenue by product category

Let's have a quick check to see if we have all product category names translated.

```python
query = """
        SELECT product_category_name
        FROM products p
        WHERE NOT EXISTS (SELECT product_category_name
                         FROM translation t
                         WHERE p.product_category_name = t.product_category_name)
        GROUP BY product_category_name
        """

select(query)
```

So there are 3 categories that don't have english translation.

```python
query = """
        SELECT order_id,
               count(distinct product_id) as num_products
        FROM order_items
        GROUP BY order_id
        ORDER BY num_products desc
        LIMIT 10
        """

select(query)
```

Since one order can have different products, I have to use both order_id and product_id in the next query. a LEFT JOIN to the translation table is also needed just in case the top category names are not translated. And again, we choose delivered orders only.

```python
query = """
        SELECT year
            ,product_category_name
            ,product_category_name_english
            ,total_revenue
        FROM
        (
            SELECT year
                ,p.product_category_name, t.product_category_name_english
                ,sum(t1.revenue) as total_revenue
                ,rank() over (partition by year order by sum(t1.revenue) desc) as value_rank
            FROM 
            (
                SELECT order_id, date_part('year', order_purchase_timestamp)::int as year
                FROM orders
                WHERE order_status = 'delivered'
            ) o
            JOIN
            (
                SELECT order_id, product_id
                    ,sum(price) + sum(freight_value) as revenue
                FROM order_items
                GROUP BY order_id, product_id
            ) t1
            on o.order_id = t1.order_id
            JOIN products p
            on t1.product_id = p.product_id
            JOIN translation t
            on p.product_category_name = t.product_category_name
            GROUP BY year, p.product_category_name, t.product_category_name_english
        ) t3
        WHERE value_rank = 1
        """

select(query)
```

The revenue of furniture_decor category, bed_bath_table category and health_beauty category is highest. 

### Canceled orders by category

```python
query = """
        SELECT year
            ,product_category_name
            ,product_category_name_english
            ,total_canceled_orders
        FROM
        (
            SELECT year
                ,p.product_category_name, t.product_category_name_english
                ,sum(t1.num_canceled_orders)::int as total_canceled_orders
                ,rank() over (partition by year order by sum(t1.num_canceled_orders) desc) as value_rank
            FROM 
            (
                SELECT order_id, date_part('year', order_purchase_timestamp)::int as year
                FROM orders
                WHERE order_status = 'canceled'
            ) o
            JOIN
            (
                SELECT order_id, product_id
                    ,count(order_id) as num_canceled_orders
                FROM order_items
                GROUP BY order_id, product_id
            ) t1
            on o.order_id = t1.order_id
            JOIN products p
            on t1.product_id = p.product_id
            JOIN translation t
            on p.product_category_name = t.product_category_name
            GROUP BY year, p.product_category_name, t.product_category_name_english
        ) t3
        WHERE value_rank = 1
        """

select(query)
```

In 2018, this category had the highest total revenue, while also the highest number of cancellation orders.
