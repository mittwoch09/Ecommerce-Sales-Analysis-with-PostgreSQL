### Checking review data

Once the customer receives the product, or the estimated delivery date is due, the customer gets a satisfaction survey by email where he can give a note for the purchase experience and write down some comments.

```python
query = """
        SELECT count(review_score) as review_score_null
        FROM reviews
        WHERE review_score = 0 or review_score is null
        """

select(query)
```

Surprisingly, there's no 0 review score and every order has at least one review score. I want to know how many scores (1 to 5) there are in each order status group.

```python
query = """
        SELECT order_status
            ,sum(case when review_score = 1 then 1 else 0 end) as score_1
            ,sum(case when review_score = 2 then 1 else 0 end) as score_2
            ,sum(case when review_score = 3 then 1 else 0 end) as score_3
            ,sum(case when review_score = 4 then 1 else 0 end) as score_4
            ,sum(case when review_score = 5 then 1 else 0 end) as score_55            
        FROM 
        (
            SELECT order_status, review_score
            FROM orders o
            JOIN reviews r on o.order_id = r.order_id
        ) a
        GROUP BY 1
        """

select(query)
```

Are these review scores assigned before or after the deadline?

```python
query = """
        SELECT order_status
            ,sum(case when 
                date_part('hour', age(review_answer_timestamp, order_estimated_delivery_date )) >= 0 
                then 1 else 0 end) as after_deadline
            ,sum(case when 
                date_part('hour', age(review_answer_timestamp, order_estimated_delivery_date)) < 0 
                then 1 else 0 end) as before_deadline
        FROM orders o
        JOIN reviews r on o.order_id = r.order_id
        GROUP BY 1
        """

select(query)
```

Now we see that customers can somehow review an order even when the deadline is not due.

And are these review scores assigned before or after the survey created?

```python
query = """
        SELECT order_status
            ,sum(case when 
                date_part('hour', age(review_answer_timestamp, review_creation_date)) >= 0 
                then 1 else 0 end) as after_survey
            ,sum(case when 
                date_part('hour', age(review_answer_timestamp, review_creation_date)) < 0 
                then 1 else 0 end) as before_survey
        FROM orders o
        JOIN reviews r on o.order_id = r.order_id
        GROUP BY 1
        """

select(query)
```

Alright, now we're sure that these review scores are collected from satisfaction surveys. And these surveys are not sent out only when the order is delivered or after the deadline is due. Well, all customers really completed these surveys thoroughly.

### Average review score

```python
query = """
        SELECT review_score
            ,count(distinct seller_id) as num_seller_by_review
        FROM reviews r
        JOIN order_items oi on r.order_id = oi.order_id
        GROUP BY 1
        ORDER BY 1
        """
select(query)
```

```python
query = """
        SELECT seller_id, avg(review_score) as avg_score
        FROM reviews r
        JOIN order_items oi on r.order_id = oi.order_id
        GROUP BY seller_id
        """
select(query)
```

```python
query = """
        SELECT b.num_sellers
            ,count(a.seller_id) as num_qualified_sellers
            ,round(count(a.seller_id) * 1.0 / b.num_sellers * 100, 2) as percentage
        FROM 
        (
            SELECT seller_id, avg(review_score) as avg_score
            FROM reviews r
            JOIN order_items oi on r.order_id = oi.order_id
            GROUP BY seller_id
            HAVING avg(review_score) >= 4
        ) a
        CROSS JOIN 
        (
            SELECT count(distinct seller_id) as num_sellers
            FROM order_items
        ) b
        GROUP BY 1
        """

select(query)
```

64.36% sellers have average review score of 4 or higher. Most customers was satisfied with their purchase, giving a rating of 4 or higher.
